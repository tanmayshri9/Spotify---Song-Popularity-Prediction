```python
def minmax_normalize_data(x):
    xmax = max(x)
    xmin = min(x)
    x_new = (x-xmin)/(xmax-xmin)
    return x_new
```

```python
# Training the Random Forest Regression model on the whole dataset
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold 
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
# from sklearn.metrics import mean_absolute_percentage_error
```

```python
forest_model = RandomForestRegressor(n_estimators=10, criterion='mae', random_state=1)
k = 5
kf = KFold(n_splits=k, random_state=None)
acc_score = list()

start_time = datetime.now().strftime("%H:%M:%S")
print("Start time =", start_time)

#model = LogisticRegression(solver= 'liblinear')
for train_index , test_index in kf.split(x_train):
    print(len(acc_score)+1)
    #print(len(train_index),len(test_index))
    x_tr , x_te = x_train[train_index,:],x_train[test_index,:]
    y_tr , y_te = y[train_index] , y[test_index]
     
    forest_model.fit(x_tr,y_tr)
    pred_values = forest_model.predict(x_te)
    acc = mean_absolute_error(y_te, pred_values)
    #acc = accuracy_score(pred_values , y_test)
    acc_score.append(acc)
    print('Finished fold "' + str(len(acc_score)) + '" at ' + datetime.now().strftime("%H:%M:%S"))
avg_acc_score = sum(acc_score)/k
 
print('accuracy of each fold - {}'.format(acc_score))
print('Avg accuracy : {}'.format(avg_acc_score))
end_time = datetime.now().strftime("%H:%M:%S")
print("End Time =", end_time)
```

```python
def percentage_error(actual, predicted):
    res = np.empty(actual.shape)
    for j in range(actual.shape[0]):
        if actual[j] != 0:
            res[j] = (actual[j] - predicted[j]) / actual[j]
        else:
            res[j] = predicted[j] / np.mean(actual)
    return res

def mape(y_true, y_pred): 
    return np.mean(np.abs(percentage_error(np.asarray(y_true), np.asarray(y_pred)))) * 100
```

```python
forest_model_new = RandomForestRegressor(n_estimators=10, criterion='mae', random_state=1)
k = 5
kf = KFold(n_splits=k, random_state=None)
mae_score = list()
mape_score = list()
mse_score = list()
rmse_score = list()

start_time = datetime.now().strftime("%H:%M:%S")
print("Start time =", start_time)


for train_index , test_index in kf.split(x_train):
    print(len(mae_score)+1)
    #print(len(train_index),len(test_index))
    x_tr , x_te = x_train[train_index,:],x_train[test_index,:]
    y_tr , y_te = y[train_index] , y[test_index]
     
    forest_model_new.fit(x_tr,y_tr)
    pred_values = forest_model_new.predict(x_te)
    
    mae = mean_absolute_error(y_te, pred_values)
    mae_score.append(mae)
    
#     mape = mape(y_te, pred_values)
#     mape_score.append(mape)
    
    mse = mean_squared_error(y_te, pred_values)
    mse_score.append(mse)
        
    rmse = np.sqrt(mean_squared_error(y_te, pred_values))
    rmse_score.append(rmse)
        
    print('Finished fold "' + str(len(mae_score)) + '" at ' + datetime.now().strftime("%H:%M:%S"))
end_time = datetime.now().strftime("%H:%M:%S")
print("End Time =", end_time)
```

```python
avg_mae_score = sum(mae_score)/k
print('MAE of each fold - {}'.format(mae_score))
print('Avg MAE : {}'.format(avg_mae_score))

# avg_mape_score = sum(mape_score)/k
# print('MAPE of each fold - {}'.format(mape_score))
# print('Avg MAPE : {}'.format(avg_mape_score))

avg_mse_score = sum(mse_score)/k
print('MSE of each fold - {}'.format(mse_score))
print('Avg MSE : {}'.format(avg_mse_score))

avg_rmse_score = sum(rmse_score)/k
print('RMSE of each fold - {}'.format(rmse_score))
print('Avg RMSE : {}'.format(avg_rmse_score))
```
